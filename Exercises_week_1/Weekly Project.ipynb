# Ex 1  & 2 Solution
1) Can you segment the apples from the tree?
2) Can you get the computer to count how many there are? How close can you get to the ground truth? (there are 26 apples in the image)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Define HSV color ranges for detecting apples
low_apple_red = (160.0, 150.0, 100.0)
high_apple_red = (180.0, 255.0, 255.0)
low_apple_raw = (0.0, 150.0, 100.0)
high_apple_raw = (15.0, 255.0, 255.0)

# Path to your image
image_path = img_path # Replace with your image path

# Read the input image
image_bgr = cv2.imread(image_path)
if image_bgr is None:
    raise FileNotFoundError("Image not found or cannot be opened.")

# Convert the image from BGR to HSV color space
image_hsv = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2HSV)

# Create masks for detecting apples
mask_red = cv2.inRange(image_hsv, low_apple_red, high_apple_red)
mask_raw = cv2.inRange(image_hsv, low_apple_raw, high_apple_raw)

# Combine masks
mask = cv2.bitwise_or(mask_red, mask_raw)

# Apply morphological operations to clean up the mask
kernel = np.ones((10, 10), np.uint8)  # Larger kernel size for better results
mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)
mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)

# Find contours in the mask
cnts, _ = cv2.findContours(mask.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Initialize apple count
c_num = 0

# Draw contours and count apples
for c in cnts:
    # Calculate the enclosing circle for each contour
    ((x, y), r) = cv2.minEnclosingCircle(c)
    if r > 8:  # Reduced threshold for better detection of smaller apples
        c_num += 1
        # Draw the circle and label on the original image
        cv2.circle(image_bgr, (int(x), int(y)), int(r), (0, 255, 0), 2)
        cv2.putText(image_bgr, "#{}".format(c_num), (int(x) - 10, int(y)), 
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 0, 0), 2)

# Convert BGR images to RGB for Matplotlib
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)
mask_rgb = cv2.cvtColor(mask, cv2.COLOR_GRAY2RGB)

# Plot the results
plt.figure(figsize=(12, 8))

plt.subplot(1, 3, 1)
plt.title('Original Image')
plt.imshow(image_rgb)
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title('Detected Apples')
plt.imshow(image_rgb)
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title('Mask Image')
plt.imshow(mask_rgb)
plt.axis('off')

plt.show()

print(f"Number of apples detected: {c_num}")

```

# Ex 3 Solution
Can you change the color of one of them?

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Define HSV color ranges for detecting apples
low_apple_red = (160.0, 150.0, 100.0)
high_apple_red = (180.0, 255.0, 255.0)
low_apple_raw = (0.0, 150.0, 100.0)
high_apple_raw = (15.0, 255.0, 255.0)

# Path to your image
image_path = img_path  # Replace with your image path

# Read the input image
image_bgr = cv2.imread(image_path)
if image_bgr is None:
    raise FileNotFoundError("Image not found or cannot be opened.")

# Convert the image from BGR to HSV color space
image_hsv = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2HSV)

# Create masks for detecting apples
mask_red = cv2.inRange(image_hsv, low_apple_red, high_apple_red)
mask_raw = cv2.inRange(image_hsv, low_apple_raw, high_apple_raw)

# Combine masks
mask = cv2.bitwise_or(mask_red, mask_raw)

# Apply morphological operations to clean up the mask
kernel = np.ones((10, 10), np.uint8)
mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)
mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)

# Find contours in the mask
cnts, _ = cv2.findContours(mask.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Color to change the apple to (e.g., blue in BGR format)
new_color = (255, 0, 0)  # BGR color (blue)

# Initialize apple count
c_num = 0

# Draw contours and change color of the first apple
for i, c in enumerate(cnts):
    if i == 0:  # Change the color of the first detected apple
        # Create a mask for the selected apple
        mask_apple = np.zeros_like(mask)
        cv2.drawContours(mask_apple, [c], -1, 255, thickness=cv2.FILLED)
        
        # Apply the new color to the selected apple
        image_bgr[mask_apple == 255] = new_color
    
    c_num += 1

# Convert BGR image to RGB for Matplotlib
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)
mask_rgb = cv2.cvtColor(mask, cv2.COLOR_GRAY2RGB)

# Plot the results
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.title('Detected Apples')
plt.imshow(image_rgb)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Mask Image')
plt.imshow(mask_rgb)
plt.axis('off')

plt.show()

print(f"Number of apples detected: {c_num}")
```

# Ex 4 Solution
Can you segment the leaves?

```python
# Define HSV color ranges for detecting darker green leaves
low_leaf_dark_green = (30.0, 40.0, 30.0)  # Adjust as needed
high_leaf_dark_green = (80.0, 255.0, 150.0)  # Adjust as needed

# Path to your image
image_path = img_path  # Replace with your image path

# Read the input image
image_bgr = cv2.imread(image_path)
if image_bgr is None:
    raise FileNotFoundError("Image not found or cannot be opened.")

# Convert the image from BGR to HSV color space
image_hsv = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2HSV)

# Create a mask for detecting darker green leaves
mask_dark_leaf = cv2.inRange(image_hsv, low_leaf_dark_green, high_leaf_dark_green)

# Apply morphological operations to clean up the mask
kernel = np.ones((10, 10), np.uint8)
mask_dark_leaf = cv2.morphologyEx(mask_dark_leaf, cv2.MORPH_CLOSE, kernel)
mask_dark_leaf = cv2.morphologyEx(mask_dark_leaf, cv2.MORPH_OPEN, kernel)

# Find contours in the mask
cnts, _ = cv2.findContours(mask_dark_leaf.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Draw contours and optionally change the color of the leaves
image_colored = image_bgr.copy()
for c in cnts:
    # Draw a contour around each detected dark green leaf
    cv2.drawContours(image_colored, [c], -1, (0, 255, 0), 2)  # Green contour for visibility

# Convert BGR image to RGB for Matplotlib
image_rgb = cv2.cvtColor(image_colored, cv2.COLOR_BGR2RGB)
mask_rgb = cv2.cvtColor(mask_dark_leaf, cv2.COLOR_GRAY2RGB)

# Plot the results
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.title('Detected Dark Green Leaves')
plt.imshow(image_rgb)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Dark Green Leaf Mask')
plt.imshow(mask_rgb)
plt.axis('off')

plt.show()
```

# Ex 4 Solution
Can you improve the edge with eroding/dilating?

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the green screen image
green_screen_image_path = pt  # Replace with your green screen image path
green_screen_img = cv2.imread(green_screen_image_path)

# Define the HSV range for green color
low_green = np.array([35, 50, 50])
high_green = np.array([85, 255, 255])

# Convert the green screen image to HSV color space
hsv_img = cv2.cvtColor(green_screen_img, cv2.COLOR_BGR2HSV)

# Create a mask for the green color
mask = cv2.inRange(hsv_img, low_green, high_green)
mask_inv = cv2.bitwise_not(mask)

# Separate the foreground (green screen)
foreground = cv2.bitwise_and(green_screen_img, green_screen_img, mask=mask_inv)

# Create a solid color background (e.g., white)
background_color = (255, 255, 255)  # BGR color (white)
background_img = np.full_like(green_screen_img, background_color)

# Combine the foreground and the new background
background = cv2.bitwise_and(background_img, background_img, mask=mask)
result = cv2.add(foreground, background)

# Convert the resulting image to grayscale
result_gray = cv2.cvtColor(result, cv2.COLOR_BGR2GRAY)

# Define a kernel for morphological operations
kernel = np.ones((5, 5), np.uint8)

# Apply dilation followed by erosion (Opening) to sharpen edges
dilated = cv2.dilate(result_gray, kernel, iterations=1)
eroded = cv2.erode(dilated, kernel, iterations=1)

# Convert the sharpened grayscale image back to BGR
sharpened = cv2.cvtColor(eroded, cv2.COLOR_GRAY2BGR)

# Convert BGR image to RGB for Matplotlib
sharpened_rgb = cv2.cvtColor(sharpened, cv2.COLOR_BGR2RGB)

# Plot the sharpened result in Jupyter Notebook
plt.figure(figsize=(8, 6))
plt.imshow(sharpened_rgb)
plt.axis('off')  # Hide axes
plt.title('Sharpened Image with Green Screen Removed')
plt.show()
```
