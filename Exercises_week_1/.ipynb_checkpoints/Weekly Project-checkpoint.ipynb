```python
bgr_img = cv2.imread("C:/Users/szymo/Desktop/DTU_Masters_1_St_semester/34759_Perception_4_Autunomous_Systems/Exercises_week_1_My_SOlutions/appletree.jpg")
b,g,r = cv2.split(bgr_img)       # get b,g,r
image = cv2.merge([r,g,b])
(h, w, d) = image.shape
print("width={}, height={}, depth={}".format(w, h, d))
```

### 1. Can you segment the apples from the tree?

```python
#HSV conversion
hsv_img = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
plt.imshow(hsv_img)
```

```python
#Defining red
lighter_red1 = np.array([0, 100, 100])     
darker_red1 = np.array([10, 255, 255])    
lighter_red2 = np.array([160, 100, 100])   
darker_red2 = np.array([180, 255, 255])   

#Defining masks
mask1 = cv2.inRange(hsv_img, lighter_red1, darker_red1)
mask2 = cv2.inRange(hsv_img, lighter_red2, darker_red2)
mask = cv2.bitwise_or(mask1, mask2)

segment_img = cv2.bitwise_and(bgr_img, bgr_img, mask=mask)

segment_img_rgb = cv2.cvtColor(segment_img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 5))

plt.title("Original Image")
plt.imshow(cv2.cvtColor(bgr_img, cv2.COLOR_BGR2RGB))
plt.show()

plt.title("Segmented Apples")
plt.imshow(segment_img_rgb)

plt.show()
```

## 2. How many apples appear there?

```python
#Morphological operations to make counting more relevant
kernel = np.ones((5, 5), np.uint8)  # Define a kernel for morphology
mask_cleaned = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)  #Fill small holes
mask_cleaned = cv2.morphologyEx(mask_cleaned, cv2.MORPH_OPEN, kernel)  #Remove noise

contours, _ = cv2.findContours(mask_cleaned, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

extracted_apples = []

#Iterating through the contours
for cnt in contours:
    area = cv2.contourArea(cnt)
    if area > 100:  #Filter using the noise-avoidant area
        x, y, w, h = cv2.boundingRect(cnt)  #Defininge boarders where the apples are placed
        apple = bgr_img[y:y+h, x:x+w]  #Apple extraction
        extracted_apples.append(apple)  # Add to list

plt.figure(figsize=(15, 10))
plt.subplot(1, 3, 1)
plt.title("Original Image")
plt.imshow(cv2.cvtColor(bgr_img, cv2.COLOR_BGR2RGB))

plt.subplot(1, 3, 2)
plt.title("Improved Mask")
plt.imshow(mask_cleaned, cmap='gray')

# Display each extracted apple
plt.subplot(1, 3, 3)
plt.title("Extracted Apples")
for i, apple in enumerate(extracted_apples):
    plt.imshow(cv2.cvtColor(apple, cv2.COLOR_BGR2RGB))  
    plt.show()

# Optionally, display count of apples
print("Number of apples extracted:", len(extracted_apples))
```

## 3.Change one's apple color

```python
max_area = 0
distinct_apple_contour = None

for cnt in contours:
    area = cv2.contourArea(cnt)
    if area > 200 and area < 1000:  #I won't be using overlapping apples for this exercise, so these are boundaries to avoid them
        if area > max_area:
            max_area = area
            distinct_apple_contour = cnt

# Check if a distinct apple was found
if distinct_apple_contour is not None:
    # Create a mask for the distinct apple
    apple_mask = np.zeros_like(mask_cleaned)
    cv2.drawContours(apple_mask, [distinct_apple_contour], -1, 255, thickness=cv2.FILLED)
    
    #Changing it color
    bgr_img_with_changed_color = bgr_img.copy()
    bgr_img_with_changed_color[apple_mask == 255] = [0, 255, 50] 

    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.title("Original Image")
    plt.imshow(cv2.cvtColor(bgr_img, cv2.COLOR_BGR2RGB))

    plt.subplot(1, 2, 2)
    plt.title("Distinct Apple Changed Color")
    plt.imshow(cv2.cvtColor(bgr_img_with_changed_color, cv2.COLOR_BGR2RGB))
    plt.show()
```

## 4. Leaves segmentation

```python
lower_green = np.array([40, 50, 50])
upper_green = np.array([90, 255, 255])

mask_green = cv2.inRange(hsv_img, lower_green, upper_green)

kernel = np.ones((5, 5), np.uint8)
mask_green_cleaned = cv2.morphologyEx(mask_green, cv2.MORPH_CLOSE, kernel)
mask_green_cleaned = cv2.morphologyEx(mask_green_cleaned, cv2.MORPH_OPEN, kernel)

contours_green, _ = cv2.findContours(mask_green_cleaned, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

leaf_segmented_img = np.zeros_like(bgr_img)

for cnt in contours_green:
    area = cv2.contourArea(cnt)
    if area > 100:
        cv2.drawContours(leaf_segmented_img, [cnt], -1, (0, 255, 0), thickness=cv2.FILLED)

plt.title("Original image")
plt.imshow(cv2.cvtColor(bgr_img, cv2.COLOR_BGR2RGB))
plt.show()

plt.title("Isolated leaves")
plt.imshow(cv2.cvtColor(leaf_segmented_img, cv2.COLOR_BGR2RGB))
plt.show()
```
