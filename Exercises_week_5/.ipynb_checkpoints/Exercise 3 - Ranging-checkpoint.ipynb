# Ranging and Ransac

We'll use the following imports:

```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn import linear_model
```

If you get an error from importing sklearn, then you will have to install it using:
conda install scikit-learn

For the exercise we will be using the data taken from a lidar by a robot in a hallway. The data is saved in `laser.csv` and we can import this in Python using numpy as such:

```python
laser = np.genfromtxt('laser.csv', delimiter=',')
print(laser)
```

The file contains only a single rotation of the lidar. You can open the file in any text editor to get a look at the data. Each number in the file is the distance measured in meters. The lidar starts measuring at 135 degrees and ends at -135 degrees and it rotates the same amount between each measurement. 

## Exercise 3.1
Convert the data in the csv file to x and y coordinates in a Carthesian coordinate system and plot the result using matplotlib.

Hint: Use [scatter](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html) instead of plt.plot in order to show every measurement as a point. 

```python
# Convert to x and y coordinates
num_msr = len(laser)
angles = np.linspace(135, -135, num_msr)

angles_1 = np.radians(angles)

x = laser * np.cos(angles_1)
y = laser * np.sin(angles_1)

plt.scatter(x,y)
```

## Exercise 3.2

We would like to locate one of the walls in the dataset. To do so we will use [Ransac](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.RANSACRegressor.html) to fine the best line fit in the noisy data. A Ransac regressor is initialized the following way using sklearn:

```python
ransac = linear_model.RANSACRegressor()
```

Now that we have our Ransac regressor we can fit it to a dataset (x,y):

```python
x = np.array(x).reshape(-1, 1)
y = np.array(y)
```

```python
ransac = linear_model.RANSACRegressor()
ransac.fit(x, y)
```

```python
ransac.score(x, y)
```

Finally, to get the fitted line we can use the predict function. As an argument this takes the x values in which the line will be predicted:

```python
ransac.predict(x[:-1])
```

Plot the line on top of the scatterplot from before to see the fit.

## Exercise 3.3

To see which data points were used to make the line fit and which were deemed outliers, we can use the function ransac.inlier_mask_

```python
plt.scatter(x[ransac.inlier_mask_], y[ransac.inlier_mask_], color='green')
```

Find the second hallway wall using another Ransac fit and plot it together with the data points and the first ransac fit.

Hint: Use the outliers from the first Ransac.

```python
ransac2 = linear_model.RANSACRegressor()

# Find the second hallway wall
inlier_mask2 = ransac.inlier_mask_
outliers_x = x[~inlier_mask2]
outliers_y = y[~inlier_mask2]

#print("Outliers (X):", outliers_x.flatten())
#print("Outliers (Y):", outliers_y.flatten())
outliers_x = np.array(outliers_x).reshape(-1, 1)
outliers_y = np.array(outliers_y)

ransac2 = linear_model.RANSACRegressor()
ransac2.fit(outliers_x, outliers_y)

ransac2.score(outliers_x, outliers_y)

ransac2.predict(outliers_x[:-1])
```

```python
# Assuming x and y are your data points
x = np.array(x).reshape(-1, 1)  # Reshape for sklearn
y = np.array(y)

# Step 1: First RANSAC fit
ransac1 = linear_model.RANSACRegressor(random_state=0)
ransac1.fit(x, y)
inlier_mask1 = ransac1.inlier_mask_

# Identify outliers from the first fit
outliers_x1 = x[~inlier_mask1]
outliers_y1 = y[~inlier_mask1]

# Step 2: Second RANSAC fit on outliers from the first fit
ransac2 = linear_model.RANSACRegressor(random_state=0)
ransac2.fit(outliers_x1, outliers_y1)

# Get the inlier mask for the second fit
inlier_mask2 = ransac2.inlier_mask_

# Identify second wall inliers and outliers
second_wall_inliers_x = outliers_x1[inlier_mask2]
second_wall_inliers_y = outliers_y1[inlier_mask2]
second_wall_outliers_x = outliers_x1[~inlier_mask2]
second_wall_outliers_y = outliers_y1[~inlier_mask2]

# Prepare data for plotting
line_x1 = np.linspace(x.min(), x.max(), 100).reshape(-1, 1)
line_y1 = ransac1.predict(line_x1)

line_x2 = outliers_x1
line_y2 = ransac2.predict(outliers_x1)

# Plotting
plt.figure(figsize=(10, 6))

# Plot first RANSAC fit
plt.scatter(x[inlier_mask1], y[inlier_mask1], color='green', label='First Wall Inliers')
plt.scatter(outliers_x1, outliers_y1, color='red', label='First Wall Outliers')
plt.plot(line_x1, line_y1, color='green', label='First Wall Fit')

# Plot second RANSAC fit
plt.scatter(second_wall_inliers_x, second_wall_inliers_y, color='blue', label='Second Wall Inliers')
plt.scatter(second_wall_outliers_x, second_wall_outliers_y, color='orange', label='Second Wall Outliers')
plt.plot(outliers_x1, line_y2, color='blue', label='Second Wall Fit')

# Finalizing the plot
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Hallway Wall Detection with RANSAC')
plt.legend()
plt.grid()
plt.show()
```
